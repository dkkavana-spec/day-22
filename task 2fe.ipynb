```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'Age': [25, 32, 47, 51, 28],
    'Salary': [50000, 65000, 120000, 98000, 55000]
})
scaler_std = StandardScaler()
data_std = pd.DataFrame(scaler_std.fit_transform(data), columns=data.columns)
scaler_minmax = MinMaxScaler()
data_minmax = pd.DataFrame(scaler_minmax.fit_transform(data), columns=data.columns)
fig, axs = plt.subplots(3, 1, figsize=(6, 6))
axs[0].hist(data['Salary'], bins=10, color='blue')
axs[0].set_title('Original Salary')
axs[1].hist(data_std['Salary'], bins=10, color='green')
axs[1].set_title('Standardized Salary')
axs[2].hist(data_minmax['Salary'], bins=10, color='red')
axs[2].set_title('Normalized Salary')
plt.tight_layout()
plt.show()
```
