```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
np.random.seed(42)
X = np.random.rand(100, 1) * 10     
y = 3 * X.squeeze()**2 + 2 * X.squeeze() + np.random.randn(100) * 3  
X_orig = X
y_target = y
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_orig)
model_orig = LinearRegression()
model_orig.fit(X_orig, y_target)
y_pred_orig = model_orig.predict(X_orig)
r2_orig = r2_score(y_target, y_pred_orig)
model_poly = LinearRegression()
model_poly.fit(X_poly, y_target)
y_pred_poly = model_poly.predict(X_poly)
r2_poly = r2_score(y_target, y_pred_poly)
print(f"R² with original features: {r2_orig:.4f}")
print(f"R² with polynomial features: {r2_poly:.4f}")
if r2_poly > r2_orig:
    print("The 'curved' (polynomial) features improved the model fit.")
else:
    print("The polynomial features did not improve the fit.")
```
